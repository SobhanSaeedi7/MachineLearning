```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from scipy import stats

from lls import LLS
```

Generate continuous random dataset for the problem of X = study hours and Y = grade of students 

```python
impact_factor = 3
study_hours = np.random.normal(3, 2, 100)
natural_mental = np.random.normal(1, 1, 100)
grade = study_hours * impact_factor + natural_mental
grade[grade > 20] = 20
grade[grade < 0] = 0
X = np.around(study_hours,1)
Y = np.around(grade,1)
```

Split your dataset to train and test with train_test_split function.


```python
X_train, X_test, Y_train, Y_test = train_test_split(study_hours, grade, test_size=0.2)
X_train = X_train.reshape(-1, 1)
```

___

```python
plt.scatter(X_train, Y_train, c = 'orange', marker='.')
plt.title('Grades of students according to their studying time')
plt.xlabel('Hours of study')
plt.ylabel('Grade')
plt.show()
```

___

Draw some wrong red lines with a random slope.

```python
wrong_w = np.random.normal(3, 1, 3)
```

```python
plt.scatter(X_train, Y_train, c = 'orange', marker='.')
plt.plot(X_train, wrong_w[0] * X_train, color='red')
plt.plot(X_train, wrong_w[1] * X_train, color='red')
plt.plot(X_train, wrong_w[2] * X_train, color='red')
plt.title('rades of students according to their studying time')
plt.xlabel('Hours of study')
plt.ylabel('Grade')
plt.legend(['data','wrong lines'])
plt.show()
```


Draw the correct blue line using your Object Oriented Linear Least Squares (LLS) method.



```python
lls = LLS()
lls_w = lls.fit(X_train, Y_train)
```

```python
plt.scatter(X_train, Y_train, c = 'orange', marker='.')
plt.plot(X_train, lls_w * X_train, color='blue')
plt.title('rades of students according to their studying time')
plt.xlabel('Hours of study')
plt.ylabel('Grade')
plt.legend(['data','LLS line'])
plt.show()
```


Draw a green line using linregress from scipy library. Compare your result with scipy's result.

```python
slope, intercept, r, p, std_err = stats.linregress(X_train.T, Y_train)
```

```python
plt.scatter(X_train, Y_train, c = 'orange', marker='.')
plt.plot(X_train, slope * X_train + intercept, color='green')
plt.title('Grades of students according to their studying time')
plt.xlabel('Hours of study')
plt.ylabel('Grade')
plt.legend(['data','scipy line'])
plt.show()
```

### Final Result

```python
plt.scatter(X_train, Y_train, c = 'orange', marker='.')
plt.plot(X_train, lls_w * X_train, color='blue', alpha=0.5)
plt.plot(X_train, slope * X_train + intercept, color='green', alpha=0.5)
plt.plot(X_train, wrong_w[0] * X_train, color='red', alpha=0.5)
plt.plot(X_train, wrong_w[1] * X_train, color='red', alpha=0.5)
plt.plot(X_train, wrong_w[2] * X_train, color='red', alpha=0.5)
plt.title('Grades of students according to their studying time')
plt.xlabel('Hours of study')
plt.ylabel('Grade')
plt.legend(['data','LLS line', 'scipy line', 'wrong lines'])
plt.show()
```
