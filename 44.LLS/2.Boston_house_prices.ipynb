```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

from scipy import stats
from sklearn.datasets import load_boston
from lls import LLS
```

Load Boston dataset from the scikit-learn library.



```python
boston = load_boston()
df = pd.DataFrame(boston.data, columns=boston.feature_names)
df['prices'] = boston.target
df.head(5)
```

According to data correlation and your analysis, choose 2 features for X and choose 1 target for Y


```python
corr = df.corr()
corr
```

Split the dataset to train and test with train_test_split function.


```python
X = df[['ZN', 'RM']].copy()
Y = df['prices'].reset_index()
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
X_train
```

Solve the regression problem on 2 dimensions (2 features) using your Object Oriented LLS.

```python
lls = LLS()
w = lls.fit(X_train, Y_train)
w = w['prices'].to_numpy()
w
```


Show results on a 3D plot and draw the y_pred as a plane on it.

```python
a, b = np.meshgrid(X_train['ZN'], X_train['RM'])
eq = w[0] * a + w[1] * b

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(X_train['ZN'] , X_train['RM'] , Y_train['prices'], color='y', alpha=0.5)
ax.plot_surface(a, b, eq, color='b', alpha=0.1)
ax.set_xlabel('RM')
ax.set_ylabel('ZN')
ax.set_zlabel('Price')
plt.show()
```
