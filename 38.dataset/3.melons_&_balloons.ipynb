```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
melon_width = np.random.normal(12, 1, 250)
melon_length = np.random.normal(12, 1, 250)
melon_weight = np.random.normal(2, 0.5, 250)
```

```python
balloon_width = np.random.normal(15, 1, 250)
balloon_length = np.random.normal(25, 1, 250)
balloon_weight = np.random.normal(0.2, 0.2, 250)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(melon_length, melon_width, melon_weight,
           color='green', alpha=0.5)

ax.scatter(balloon_length, balloon_width, balloon_weight,
           color='red', alpha=0.5)

plt.title('Melons & Ballons')
ax.set_xlabel("width 'cm'")
ax.set_ylabel("length 'cm'")
ax.set_zlabel("weight 'kg'")

ax.view_init(30, 5)

plt.show()
```
