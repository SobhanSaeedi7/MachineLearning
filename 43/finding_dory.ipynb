```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

from knn import KNN
```

```python
fish = cv2.imread('Inputs/tangblue1.jpg')
fish = cv2.resize(fish, (0,0), fx=0.5, fy=0.5)
fish = cv2.cvtColor(fish, cv2.COLOR_BGR2RGB)
plt.imshow(fish)
plt.show()
```

```python
pixels_list_rgb = fish.reshape(-1, 3)

pixels_list_rgb.shape

fig = plt.figure(figsize=(8, 8))
ax =  fig.add_subplot(projection='3d')

ax.scatter(pixels_list_rgb[:, 0], pixels_list_rgb[:, 1], pixels_list_rgb[:, 2], c=pixels_list_rgb / 255)
ax.set_xlabel('Red')
ax.set_ylabel('Green')
ax.set_zlabel('Blue')
plt.show()
```

```python
fish_hsv = cv2.cvtColor(fish, cv2.COLOR_BGR2HSV)
```

```python
pixels_list_hsv = fish_hsv.reshape(-1, 3)

pixels_list_hsv.shape

fig = plt.figure(figsize=(8, 8))
ax =  fig.add_subplot(projection='3d')

ax.scatter(pixels_list_hsv[:, 0], pixels_list_hsv[:, 1], pixels_list_hsv[:, 2], c=pixels_list_rgb / 255)
ax.set_xlabel('Hue')
ax.set_ylabel('Saturation')
ax.set_zlabel('Value')
plt.show()
```

```python
light_blue = (0, 50, 100)
dark_blue = (30, 255, 255)

mask_blue = cv2.inRange(fish_hsv, light_blue, dark_blue)
plt.imshow(mask_blue, cmap='gray')
plt.show()
```

```python
light_yellow = (80, 140, 100)
dark_yellow = (100, 255, 255)

mask_yellow = cv2.inRange(fish_hsv, light_yellow, dark_yellow)
plt.imshow(mask_yellow, cmap='gray')
plt.show()
```

```python
light_black = (0, 0, 0)
dark_black = (255, 255, 90)

mask_black = cv2.inRange(fish_hsv, light_black, dark_black)
plt.imshow(mask_black, cmap='gray')
plt.show()
```

```python
mask = mask_blue + mask_yellow + mask_black
plt.imshow(mask, cmap='gray')
plt.show()
```

```python
result = cv2.bitwise_and(fish, fish, mask=mask)
plt.imshow(result, cmap='gray')
plt.show()
```
