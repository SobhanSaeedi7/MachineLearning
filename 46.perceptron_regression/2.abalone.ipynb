```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split

from perceptron import Perceptron
```

Fit your perceptron algorithm on the abalone dataset


```python
data = pd.read_csv("Inputs/abalone.csv")
data.head()
```

```python
X = data[['Length']].values
Y = data[['Diameter']].values

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
perceptron = Perceptron()
perceptron.fit(X_train, Y_train)
```

```python
print(perceptron.evaluate(X_test, Y_test))
print(perceptron.evaluate(X_test, Y_test, metric='mse'))
print(perceptron.evaluate(X_test, Y_test, metric='rmse'))
```

Plot data graph and loss graph as 2 subplots in 1 window


```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

Y_pred = perceptron.predict(X_train)

ax1.scatter(X_train, Y_train)
ax1.plot(X_train, Y_pred, color='red')
ax1.set_xlabel('Diameter');ax1.set_ylabel('Length')
ax1.set_title('Abalone')


ax2.plot(perceptron.losses)
ax2.set_title("Loss graph")
ax2.set_xlabel("Epochs")
ax2.set_ylabel("Losses")

plt.show()
```

Change hyperparameters for get best result

```python
perceptron = Perceptron(lrw=0.01, lrb=0.01, epochs=50)
perceptron.fit(X_train, Y_train)
```

```python
print(perceptron.evaluate(X_test, Y_test))
print(perceptron.evaluate(X_test, Y_test, metric='mse'))
print(perceptron.evaluate(X_test, Y_test, metric='rmse'))
```
