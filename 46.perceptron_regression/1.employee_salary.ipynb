```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

from sklearn.datasets import make_regression
from perceptron import  Perceptron
```

Make a linear dataset using scikit-learn library

```python
x, y, coef = make_regression(n_samples=125, n_features=1, n_informative=1, noise=10, coef=True, random_state=0)

x = np.interp(x, (x.min(), x.max()), (5, 15))

y = np.interp(y, (y.min(), y.max()), (8000000, 50000000))

plt.plot(x, y, '.')
plt.xlabel('Hours of work');plt.ylabel('Salary ريال')
plt.title('Hours of work Vs. Salary')
plt.show()
```

Split data to train and test datasets

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, shuffle=True, test_size=0.2)
Y_train = Y_train.reshape(-1, 1)
Y_test = Y_test.reshape(-1, 1)
```

Fit your model on the employee's salary dataset


```python
perceptron = Perceptron()
perceptron.fit(X_train, Y_train)
```

```python
print(perceptron.evaluate(X_test, Y_test))
print(perceptron.evaluate(X_test, Y_test, metric='mse'))
print(perceptron.evaluate(X_test, Y_test, metric='rmse'))
```

Plot data graph and loss graph as 2 subplots in 1 window


```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

Y_pred = perceptron.predict(X_train)

ax1.scatter(X_train, Y_train)
ax1.plot(X_train, Y_pred, color='red')
ax1.set_xlabel('Hours of work');ax1.set_ylabel('Salary ريال')
ax1.set_title('Data graph')


ax2.plot(perceptron.losses)
ax2.set_title("Loss graph")
ax2.set_xlabel("Epochs")
ax2.set_ylabel("Losses")
```

Change hyperparameters for get best result

```python
perceptron = Perceptron(lrw=0.001, epochs=15)
perceptron.fit(X_train, Y_train)
```

```python
print(perceptron.evaluate(X_test, Y_test))
print(perceptron.evaluate(X_test, Y_test, metric='mse'))
print(perceptron.evaluate(X_test, Y_test, metric='rmse'))
```
